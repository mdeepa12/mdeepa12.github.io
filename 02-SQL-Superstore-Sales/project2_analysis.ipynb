# 📊 Superstore Sales Analysis (SQL + Python)
This notebook uses pandas + matplotlib to analyze the Superstore dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (replace path if needed)
df = pd.read_csv('Superstore.csv')
df.head()
```

```python
# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Year'] = df['Order Date'].dt.year

# Sales & Profit by Year
sales_by_year = df.groupby('Year')[['Sales','Profit']].sum().reset_index()
sales_by_year
```

```python
# Plot Sales by Year
plt.figure(figsize=(8,5))
plt.bar(sales_by_year['Year'].astype(str), sales_by_year['Sales'], color='skyblue')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```
